```python
# ë‹¤ìš´ë¡œë“œí•  W&B í”„ë¡œì íŠ¸ì™€ ì—”í‹°í‹° ì„¤ì •
entity = "zerobeak"  # ì‚¬ìš©ì ë˜ëŠ” ì¡°ì§ ì´ë¦„
project = "base_v2"  # í”„ë¡œì íŠ¸ ì´ë¦„

# API ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
api = wandb.Api()

# í”„ë¡œì íŠ¸ ë‚´ ëª¨ë“  ëŸ° ì •ë³´ ê°€ì ¸ì˜¤ê¸°
runs = api.runs(f"{entity}/{project}")

# ë‹¤ìš´ë¡œë“œ í´ë” ì„¤ì •
download_path = "./videos"

# ë‹¤ìš´ë¡œë“œ í´ë”ê°€ ì—†ì„ ê²½ìš° ìƒì„±
if not os.path.exists(download_path):
    os.makedirs(download_path)

# ëª¨ë“  ëŸ°ì—ì„œ íŒŒì¼ ë‹¤ìš´ë¡œë“œ
for run in runs:
    # ê° ëŸ°ë§ˆë‹¤ artifact ê°€ì ¸ì˜¤ê¸°
    for file in run.files():
        file_path = os.path.join(download_path, run.name, file.name)

        # íŒŒì¼ì„ ì €ì¥í•  ë””ë ‰í† ë¦¬ê°€ ì—†ë‹¤ë©´ ìƒì„±
        os.makedirs(os.path.dirname(file_path), exist_ok=True)

        # íŒŒì¼ ë‹¤ìš´ë¡œë“œ
        file.download(root=os.path.join(download_path, run.name), exist_ok=True)

print("ë‹¤ìš´ë¡œë“œ ì™„ë£Œ!")
```
